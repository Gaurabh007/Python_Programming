#### iloc() and loc()
* Both method extracts the rows or column from the dataset.
* Difference: -
>* .iloc() (inbuilt/integer) - Based on the inbuilt location it extracts the rows or column. It ignores the upper-bound during selection.
>* loc() (named) - Based on the named indexes it use to extract the data. It considers the upper-bound during selection.

```python
df.iloc[0:2]
```

```python
df.loc[0:2]
```

```python
df.iloc[0:2,['PassengerId','Survived','Pclass']]    #it will throw a error because it consider only the inbuilt indexes
```

```python
df.iloc[0:2,[0,1,2,3]]
```

```python
df.loc[0:2,['PassengerId','Survived','Pclass']]
```

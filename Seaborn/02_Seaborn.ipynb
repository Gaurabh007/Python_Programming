```python
import seaborn as sns
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

## 4. Categorical Plots

### Categorical Scatter Plot (Bivariate Analysis)
- Stripplot
- Swarmplot
### Categorical Distribution Plots (Univariate Analysis)
- Boxplot
- Violinplot
### Categorical Estimate Plot -> Central Tendency
- Barplot
- Pointplot
- Countplot

### 4.1 Strip Plot -> Axes level Function
- It plots a stripplot between a categorical and numerical data.
- By default seaborn adds some noise in each category for visualization purpose

```python
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
# We can remove so by making the jitter parameter as False
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

#### Catplot -> Figure Level Function
.catplot(): When we have a Categorical data in our dataset and want to visualize with other columns of your dataset like numerical dataset .catplot() is very helpful.

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='strip')
```

```python
#  jitter parameter
sns.stripplot(data=tips,x='day',y='total_bill',jitter=0.3,hue='sex')
```

### 4.2 Swarmplot -> Axes level function
- It is similar to the strip plot
- The extra information that it tells is that it tells about the distribution of the data or how the data is divided.

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm',hue='sex')
```

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex')
```

### 4.3 Box Plot -> Axes level function
A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
# using catplot
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# single boxplot -> numerical column
sns.boxplot(data=tips,y='total_bill')
```

### 4.4 Violin Plot (Boxplot + KDE Plot)

```python
# violin plot
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)   #side by side representation
```

### 4.5 Bar Plot

```python
# This plot gives the average total bill for both the gender.
# It uses the average estimator to find the total bill value
sns.barplot(data=tips,x='sex',y='total_bill')
```

```python
# We can change the estimator value by using the estimator parameter
import numpy as np
sns.barplot(data=tips,x='sex',y='total_bill',estimator=np.median)
```

```python
# hue
sns.catplot(data=tips,x='sex',y='total_bill',kind='bar',hue='smoker')
```

### 4.6 Point Plot
- A point plot represents an estimate of central tendency for a numeric variable by the position of the dot and provides some indication of the uncertainty around that estimate using error bars.
- Its main focus is on the difference between the category.

```python
# pointplot
sns.pointplot(data=tips, x='sex',y='total_bill',errorbar=None)
```

```python
sns.barplot(data=tips,x='sex',y='total_bill')
```

### 4.7 Count Plot
- Show the counts of observations in each categorical bin using bars.
- A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable.

```python
sns.countplot(data=tips,x='sex')
```

```python
sns.countplot(data=tips,x='sex',hue='day')
```

```python
sns.catplot(data = tips,x='sex',y='total_bill',col='smoker',kind='box',row='time')
```

.load_dataset(): Seaborn has some inbuilt dataset within it and we can access it through this functin.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
tips = sns.load_dataset('tips')

tips.head()
```

## 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot
### 1.1 Scatter plot ->  Axes Level Function
Draw a scatter plot with possibility of several semantic groupings.

```python
sns.scatterplot(data = tips,x ='total_bill',y = 'tip',hue='size',style='time',size='size')
```

```python
# relplot -> figure level function -> square shape (diff. b/w scatterplot and relplot)
# facet plot -> figure level function -> works with relplot as compared to scatterplot and lineplot
sns.relplot(data = tips,x ='total_bill' , y = 'tip',kind = 'scatter',hue='day',col='smoker',row='sex')
```

Now the question arises which one to use scatterplot or relplot?
- It is adviced by Seaborn Library to use relplot (figure level) as because we get and advantage of figure level and we can replicate multiple axes level function by just changing the `kind` parameter.

```python
# Example of using relplot
sns.relplot(data = tips,x='total_bill',y='tip',kind='line')
```

### 1.2. Line Plot -> Axes Level Function

```python
# line plot
gap = px.data.gapminder()       #dataset
df = gap[gap['country'] == 'India']
df
```

```python
sns.lineplot(data=df,x='year',y='gdpPercap')
```

```python
sns.relplot(data=df,x='year',y='gdpPercap',kind='line')
```

```python
temp_df = gap[gap['country'].isin(['India','China','Sri Lanka'])]
temp_df
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line',hue='country')
```

```python
continent_df = gap[gap['country'].isin(['India','Germany','Brazil'])]
continent_df
```

```python
sns.relplot(data=continent_df,x='year',y='lifeExp',kind='line',hue='country',style='continent')
```

```python
# col_wrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```

## 2. Distribution Plots

- Used for univariate analysis
- Used to find out the distribution
- Range of the observation
- Central Tendency
- Is the data bimodal?
- Are there outliers?

Plots under distribution plot

- Histplot
- Kdeplot
- Rugplot

### histplot(): It creates a histogram of a particular column from the dataset. (figure level)
### displot() -> axes level function

```python
# displot
sns.displot(tips['total_bill'],kind='hist')      #(figure level function)
```

```python
# histplot
sns.histplot(data = tips['total_bill'])      #(axes level function)
```

```python
# bins parameter
sns.displot(data = tips['total_bill'],bins=20) 
```

```python
sns.displot(data = tips['total_bill'],bins=2) 
```

```python
# It's possible to plot a categorical data in a histplot
sns.displot(data = tips['day']) 
```

```python
# hue parameter
sns.displot(data = tips,x='tip',hue='sex') 
```

```python
# element parameter -> It gives a quite clear visuals
sns.displot(data = tips,x='tip',hue='sex',element='step') 
```

```python
titanic = sns.load_dataset('titanic')
sns.displot(data=titanic,x='age',hue='sex',element='step')      #another example of element parameter
```

### Kdeplot -> It is used to *smoothes* the observation with a Gaussian kernal, producing a continuous density estimate.

```python
tips.head()
```

```python
sns.kdeplot(data=tips,x='total_bill')   
# y-axis represents the density of probability of the data
```

```python
# We can do the kdeplot through displot()
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue and fill parameter
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

### Rugplot() -> Plot marginal distribution by drawing along the x and y axes.
### It is generally used with some other plots to show some extra add-on features

```python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

## Bivariate Histogram
### A bivariate histogram bins the data within rectangles that tile the plot and then shows the count of observations within each rectangle with the fill color

```python
sns.histplot(data=tips,x='total_bill',y='tip')
```

## Bivariate Kdeplot
### a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian

```python
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

## 3. Matrix Plot

- Heatmap
- Clustermap

### 3.1 HeatMap -> Plot rectangular data as a color-encoded matrix (axes level function)

```python
gap.head()          #Dataset
```

```python
life_exp = gap.pivot(index='country',columns='year',values='lifeExp')
life_exp
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(life_exp)
```

```python
# annot
Eup = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(10,10))
sns.heatmap(Eup,annot=True)
```

```python
# linewidth
plt.figure(figsize=(10,10))
sns.heatmap(Eup,annot=True,linewidths=0.5)
```

```python
# cmap
plt.figure(figsize=(10,10))
sns.heatmap(Eup,annot=True,linewidths=0.5,cmap='summer',linecolor='red')
```

### 3.2 Cluster Map ->  Plot a matrix dataset as a hierarchically-clustered heatmap.
- It maps the similar columns and disjoints the disimilar columns.

```python
iris = px.data.iris()
iris.head()
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
# Firstly it rearange the columns and maps the two most similar column and does the further mapping (here 'sepal width','petal length').
# It performs the mapping on both columns and rows.
# It follow a Machine Learning Algorithm i.e., Hierarchical Agglomerative Clustering.
```

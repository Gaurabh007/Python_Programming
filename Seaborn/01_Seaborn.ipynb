## Python - Seaborn Library

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

### scatterplot: Draw a scatter plot with possibility of several semantic groupings.

```python
sns.scatterplot(x = df.petal_length , y = df.sepal_length)
```

```python
sns.scatterplot(x = df.sepal_width , y = df.petal_width)
```

### displot(): It creates a histogram of a particular column from the dataset.

```python
sns.displot(df['sepal_length'])
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
import pandas as pd
tips.to_csv('tips.csv')
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip)
```

### .replot(): It is a groupby operation, in which `hue` attribute is used to differentiate the dataset based on it.

```python
sns.relplot(x = tips.total_bill, y = tips.tip , data = tips, hue = 'smoker')
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip , data = tips, hue = 'size')
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip , data = tips, style = 'size')
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip , data = tips, style = 'size',hue = 'time')
```

### .catplot(): When we have a Categorical data in our dataset and want to visulize with other columns of your dataset like numerical dataset .catplot() is very helpful.

```python
sns.catplot(y ='total_bill', x = 'day',data = tips)
```

```python
sns.catplot(x = tips.smoker, y = tips.tip)
```

### .joinplot() : It simply a combination of scatter plot and histogram. 

```python
sns.jointplot(x= tips.total_bill,y=tips.tip)
```

### .pairplot(): To plot a multiple bivariate distributions in a dataset.

```python
import matplotlib.pyplot as plt
sns.pairplot(tips,hue='time')
plt.show()
```

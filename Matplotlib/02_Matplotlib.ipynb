```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Colouring Scatter Plots

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris['species'].value_counts()
```

```python
iris['species'] = iris['species'].replace({'setosa':0,'versicolor':1,'virginica':2})
iris
```

```python
# cmap -> Theme

plt.scatter(iris['sepal_length'],iris['petal_length'],c=iris['species'],cmap='plasma')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.colorbar()
```

## Annotation

```python
df = pd.read_csv('batter.csv')
sample_df = df.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.figure(figsize=[18,10])
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

## Vertical and Horizontal Lines

```python
plt.figure(figsize=[18,10])
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

## Subplots
* We can make Multiple Graphs in a single figure
* It creates two things i.e., Figure and axes

```python
# For labeling the graph it is slightly different from the plt.scatter
fig, ax = plt.subplots(figsize=[15,6])
ax.scatter(df['avg'],df['strike_rate'])
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
ax.set_title('Average vs Strike Rate')
```

```python
# WAY - 1
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=[15,10])      #here 2 axes are there in array form and 1 figure

ax[0].scatter(df['avg'],df['strike_rate'],color = 'green')
ax[1].scatter(df['avg'],df['runs'])

ax[0].set_title('Average vs Strike Rate')
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Average vs Runs')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Run')
```

```python
# WAY - 2 (*)
# Plotting 4 Axes in 1 figure
fig,ax = plt.subplots(nrows=2,ncols=2)

ax[0,0].scatter(df['avg'],df['strike_rate'],color = 'green')
ax[0,1].scatter(df['avg'],df['runs'])
ax[1,0].hist(df['runs'])
ax[1,1].hist(df['avg'],color='green')
```

```python
# WAY - 3
fig = plt.figure(figsize=[10, 7])
ax0 = fig.add_subplot(2, 1, 1)
ax0.scatter(df['avg'], df['strike_rate'], color='green')

ax1 = fig.add_subplot(2, 2, 4)
ax1.hist(df['avg'])

plt.show()
```

## 3D-Scatter Plot

```python

fig = plt.figure(figsize=[10,6])

ax = plt.subplot(projection = '3d')

ax.scatter3D(df['runs'],df['strike_rate'],df['avg'])

ax.set_title('IPL Analysis')
ax.set_xlabel('Runs')
ax.set_ylabel('Average')
ax.set_zlabel('Strike Rate')
```

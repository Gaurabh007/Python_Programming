## Python - Matplot Library

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np 
```

```python
x = np.random.randint(1,50,(10,5))
y = np.random.randint(1,50,(10,5))
```

```python
x
```

```python
y
```

### xlabel() & ylabel(): Naming the x and y label.

```python
plt.scatter(x,y)
# Labeling the X and Y Axis
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# Labels title
plt.title('X v/s Y')
```

## Scatter Plot
* Bivariate Analysis
* Numerical vs Numerical

### scatter(): It plot each data points of x and y in a graph.

```python
plt.scatter(x,y)        
```

```python
# Changing the color
plt.scatter(x,y,c='r')
```

```python
# Changing the intensity of the points.
plt.scatter(x,y,c='g',alpha=.4)     #alpha controls the intensity of the colour
```

```python
# Changing the marker
plt.scatter(x,y,c='b',marker='+')
```

```python
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('X v/s Y')
# To add a grid in the graph
plt.grid()
```

```python
# Changing the figure size of the graph
plt.figure(figsize=[6,4])
plt.scatter(x,y,c = 'y')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('X v/s Y')
```

```python
data = pd.read_csv('batter.csv')
df = data.head(50)
df
```

```python
plt.scatter(x=df['avg'],y=df['strike_rate'],color = 'Red')
plt.title('Average vs Strike rate Analysis')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
df = sns.load_dataset('tips')
df.head(2)
```

```python
# Now this graph is showing 3 columns which the total bill and the tip and one extra column which shows the number of customer. (size column is multiplied by 10 to show the clear picture as the values are very small like 2,3,1,..)
plt.scatter(df['total_bill'],df['tip'],s=df['size']*10)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total bill vs Tip')
```

```python
# scatter plot using plt.plot
plt.plot(df['total_bill'],df['tip'],'o')
```

## 2D Line Plot

```python
player = pd.read_csv('sharma-kohli.csv')
player
```

```python
# Plotting the player dataframe
plt.plot(player['index'],player['V Kohli'])
plt.plot(player['index'],player['RG Sharma'])
```

```python
# Labels Title
plt.plot(player['index'],player['V Kohli'])
plt.plot(player['index'],player['RG Sharma'])

plt.title('Rohit Sharma vs Virat Kholi')
plt.xlabel('Session')
plt.ylabel('Runs')
```

```python
# Changing the color of the Graph
plt.plot(player['index'],player['V Kohli'],color = 'Red')
plt.plot(player['index'],player['RG Sharma'],color = 'Blue')

plt.title('Rohit Sharma vs Virat Kholi')
plt.xlabel('Session')
plt.ylabel('Runs')
```

```python
# Changing the color and style of the line
plt.plot(player['index'],player['V Kohli'],color = 'Red',linestyle = 'solid',linewidth = 3)
plt.plot(player['index'],player['RG Sharma'],color = 'Blue',linestyle = 'dashdot',linewidth = 2)

plt.title('Rohit Sharma vs Virat Kholi')
plt.xlabel('Session')
plt.ylabel('Runs')
```

```python
# Changing the color and style of the line
plt.plot(player['index'],player['V Kohli'],color = 'Red',linestyle = 'solid',linewidth = 3,marker = 'D',markersize = 10)
plt.plot(player['index'],player['RG Sharma'],color = 'Blue',linestyle = 'dashdot',linewidth = 2,marker = 'o', markersize=7)

plt.title('Rohit Sharma vs Virat Kholi')
plt.xlabel('Session')
plt.ylabel('Runs')
```

```python
# Legend -> Location
plt.plot(player['index'],player['V Kohli'],color = 'Red',linestyle = 'solid',linewidth = 3,marker = 'D',markersize = 10,label = 'Virat')
plt.plot(player['index'],player['RG Sharma'],color = 'Blue',linestyle = 'dashdot',linewidth = 2,marker = 'o', markersize=7,label = 'Rohit')

plt.title('Rohit Sharma vs Virat Kholi')
plt.xlabel('Session')
plt.ylabel('Runs')

plt.legend()
```

### limit function: When we want to trim our graph only what to show within the range data in such case we can use ylim and xlim function.

```python
# limited Axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(10000,100000)
plt.xlim(2017,2021)
```

```python
# Grid Function
plt.plot(player['index'],player['V Kohli'],color = 'Red',linestyle = 'solid',linewidth = 3,marker = 'D',markersize = 10,label = 'Virat')
plt.plot(player['index'],player['RG Sharma'],color = 'Blue',linestyle = 'dashdot',linewidth = 2,marker = 'o', markersize=7,label = 'Rohit')

plt.title('Rohit Sharma vs Virat Kholi')
plt.xlabel('Session')
plt.ylabel('Runs')

plt.legend()
plt.grid()
```

## Bar Plot

```python
c = ['a','b','c','d','e']
d = np.random.rand(5)
```

```python
# Categorical vs numerical bar plot
plt.bar(c,d)
plt.xlabel('Categorical Label')
plt.ylabel('Numerical Label')
plt.grid()
```

```python
# Ṛotating the Graph
plt.barh(c,d)
plt.xlabel('Categorical Label')
plt.ylabel('Numerical Label')
plt.grid()
```

## Plotting 3D Graph

```python
df1 = np.random.rand(50)
df2 = np.random.rand(50)
df3 = np.random.rand(50)

fig = plt.figure()  #A figure to plot a graph
ax = fig.add_subplot(projection='3d')   
ax.scatter(df1,df2,df3,c='g')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show()
```

#### Finding and Applying Queries 

```python
# To find a particular data in database
# We can write the query within the parentheses of .find().
for i in coll3_create.find({'product' : 'Affordable AI'}):
    print(i)
```

```python
# Query - Find the record where id >= 4
#gte: greater than equal to. Its a keyword in Mongodb.
for i in coll3_create.find({'_id':{'$gte' : '4'}}):
    print(i)

# {'_id':{'$gte' : '4'}} -> This is the query which says find the id which is greater than equal to ($gte) '4'.
```

#### Updation in Data

```python
# Write a query to change the company name from 'iNeuron' to 'PWSkills'.

coll3_create.update_many({'companyName': 'iNeuron'},{"$set" : {'companyName':'PWSkills'}})
# $set :- This command is used to set the new value with the replaced one.
```
